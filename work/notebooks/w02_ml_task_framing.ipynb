## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Type: Scoring and Ranking
Why: Because we aren't just saying "Yes" or "No" to a page. We are trying to assign a specific score to every page on the website. This allows us to rank them so the content team knows which page is the #1 priority to fix, which is #2, and so on.



## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

What I would predict: Click Uplift (Potential Clicks)
Where it comes from: This is a proxy. Since we don't know the future, we define "Opportunity" as the gap between how many people see the page (Impressions) and how many people actually click (Clicks). A page with high impressions but low clicks is a "Gold Mine" that just needs a better title or content.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Metric: Precision @ K (e.g., Precision @ 10)
What 'good' means: If my model picks the top 10 "Opportunity" pages, and the content team agrees that 8 of them really do need an update, my Precision @ 10 is 80%. In this lane, "good" means our top-ranked suggestions are actually useful to a human editor.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd

# 1. Load the starter data
# Note: The path might be different depending on your setup,
# but usually it's in the data folder.
try:
    df = pd.read_csv('../../data/search_data.csv')
except:
    # If the above fails, we'll use a dummy example so the notebook still runs
    print("Data file not found, showing structure example:")
    data = {'url': ['site.com/p1', 'site.com/p2'], 'clicks': [10, 50], 'impressions': [1000, 500]}
    df = pd.DataFrame(data)

# 2. Define Unit of Analysis: One row = One URL
unit_of_analysis = df.groupby('url').agg({
    'clicks': 'sum',
    'impressions': 'sum'
}).reset_index()

# 3. Create a simple "Opportunity Score" Column (The Target Proxy)
# Score = Impressions * (1 - CTR) -> basically "Missed Clicks"
unit_of_analysis['ctr'] = unit_of_analysis['clicks'] / unit_of_analysis['impressions']
unit_of_analysis['opportunity_score'] = unit_of_analysis['impressions'] * (1 - unit_of_analysis['ctr'])

# Display the first 5 rows
print("One row represents one unique Page (URL):")
display(unit_of_analysis.head())
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A fixed rule would be something like: "Update every page older than 12 months."
This is inefficient because:

Some old pages are "Evergreen" and still perform perfectly.
Some new pages might have a technical bug and need a refresh immediately.
ML can look at multiple factors at once (Click trends, Impression gaps, and Page age) to find the "hidden gems" that a simple human rule would miss.
